```python
removing_phrases = ['<?xml version="1.0" ?>\n', '<!DOCTYPE cwml>\n']
with open("15.xml", "r+") as f:
    d = f.readlines()
    f.truncate(0)
    f.close()    
```

```python
with open('15.xml', 'w') as f:
    for item in d:
        if item not in removing_phrases:
            f.write("%s\n" % item)
```

```python
new_list = []
for i in d:
    if i not in removing_phrases:
        new_list.append(i)
```
